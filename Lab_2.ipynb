```python
#one sample t test

import numpy as np
import scipy.stats as stats

#data
scores = [78, 72, 74, 76, 73, 79, 77, 71, 70, 74]
length = len(scores)
population_mean = 75

#statistics
stddev = np.std(scores,ddof=1)
scoremean = np.mean(scores)

#calculate statistics
t_test = (scoremean - population_mean)/(stddev/np.sqrt(length))

#degree of freedom
df = length -1
p_value = 2 * stats.t.sf(abs(t_test), df)

#result
print(f"Sample mean: {scoremean:.2f}" )
print(f"t-statistics: {t_test:.2f}")
print(f"p-value: {p_value:.4f}")

if p_value < 0.05:
  print("Reject the null hypothesis. The sample mean is slightly different")
else:
  print("Failed to reject the null hypothesis. No significant difference.")
```

```python
#two sample t test

import numpy as np
import scipy.stats as stats

class_a = [85, 88, 90, 92, 86, 84, 87, 89, 91, 93]
class_b = [78, 80, 82, 79, 81, 77, 83, 76, 75, 80]

t_stat, p_value = stats.ttest_ind(class_a, class_b, equal_var=False)

print(f"t-statistic: {t_stat:.2f}")
print(f"p-value: {p_value:.4f}")
if p_value < 0.05:
    print("Reject the null hypothesis: The means are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference.")
```

```python
import numpy as np
import scipy.stats as stats

# z test

#datas
sample_mean = 495
population_mean = 500
population_Std = 10
n = 40

#calculate  z stats

z_stat = (sample_mean - population_mean)/ (population_Std/np.sqrt(n))

# Calculate p-value (two-tailed)
p_value = 2 * stats.norm.sf(abs(z_stat))

# Results
print(f"z-statistic: {z_stat:.2f}")
print(f"pvalue: {p_value:.4f}")
if p_value < 0.05:
  print("Reject the null hypothesis: The sample mean is significantly different.")
else:
  print("Fail to reject the null hypothesis: No significant difference.")
```

```python
# two sample z test

import numpy as np
import scipy.stats as stats

#data
mean1, std1, n1 = 495, 10, 50
mean2, std2, n2 = 500, 12, 60

z_stats = (mean2 - mean1)/ ((std1**2 + n1 + std2**2/n2 + n2)) ** 0.5

#two tailed p_Value

p_value = 2 *stats.norm.sf(abs(z_stats))

print(f"z-statistic: {z_stat:.2f}")
print(f"pvalue: {p_value:.4f}")
if p_value < 0.05:
  print("Reject the null hypothesis: The means are significantly different.")
else:
  print("Fail to reject the null hypothesis: No significant difference.")
```
