Generate a population of data following an exponential distribution. Draw random samples and calculate the means. Observe how the
distribution of sample means converges to a normal distribution as you increase the sample size.

```python
import numpy as np
import matplotlib.pyplot as plt


def simulate_clt(data, sample_size, num_samples):
    sample_means = []
    for _ in range(num_samples):
        sample = np.random.choice(data, size=sample_size)
        sample_means.append(np.mean(sample))
    return sample_means



population_exp = np.random.exponential(scale=5, size=10000)

sample_means_exp = simulate_clt(population_exp, sample_size=30, num_samples=1000)


plt.figure(figsize=(10, 6))


plt.subplot(2, 1, 1)
plt.hist(population_exp, bins=50)
plt.title("Exponential Population Distribution")

plt.subplot(2, 1, 2)
plt.hist(sample_means_exp, bins=30)
plt.title("Sample Means (Exponential)")

plt.tight_layout()
plt.show()



population_binom = np.random.binomial(n=10, p=0.5, size=10000)

sample_means_binom = simulate_clt(population_binom, sample_size=30, num_samples=1000)


plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.hist(population_binom, bins=50)
plt.title("Binomial Population Distribution")

plt.subplot(2, 1, 2)
plt.hist(sample_means_binom, bins=30)
plt.title("Sample Means (Binomial)")

plt.tight_layout()
plt.show()
```

Problem: Sampling from a Binomial Distribution Scenario: The binomial distribution is commonly used to model binary outcomes, such as the
success or failure of an experiment (e.g., coin flips, customer conversions). In this case, you can apply the CLT to observe the normal
approximation ofthe binomial distribution as you take more samples.
Exercise:
Generate a population using the binomial distribution (e.g., number of successes in 10 trials with a 0.5 probability of success). Draw random
samples and compute their means. Plot the distribution of sample means and observe its convergence to normality.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def simulate_clt(data, sample_size, num_samples):
    means = []
    for _ in range(num_samples):
        sample = np.random.choice(data, size=sample_size)
        means.append(np.mean(sample))
    return means

population_binom = np.random.binomial(n=10, p=0.5, size=10000)
sample_means_binom = simulate_clt(population_binom, 30, 1000)

plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.hist(population_binom, bins=30)
plt.title("Binomial Population Distribution")

plt.subplot(2, 1, 2)
plt.hist(sample_means_binom, bins=30)
plt.title("Sampling Distribution (Binomial)")

plt.tight_layout()
plt.show()

population = np.random.uniform(0, 10, 10000)

sample_sizes = [5, 30, 50, 100]
num_samples = 1000

sample_means = {}

for size in sample_sizes:
    means = []
    for _ in range(num_samples):
        sample = np.random.choice(population, size=size)
        means.append(np.mean(sample))
    sample_means[size] = means

plt.figure(figsize=(10, 6))

for size in sample_sizes:
    sns.histplot(sample_means[size], kde=True, label=f"n = {size}", stat="density")

plt.title("CLT: Sample Means for Different Sample Sizes")
plt.xlabel("Sample Mean")
plt.ylabel("Density")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Demonstration of Central Limit Theorem using uniform distribution and varying sample sizes

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

population = np.random.uniform(0, 10, 10000)

sample_sizes = [5, 30, 50, 100]
num_samples = 1000

sample_means = {size: [] for size in sample_sizes}

for size in sample_sizes:
    for _ in range(num_samples):
        sample = np.random.choice(population, size=size, replace=False)
        sample_means[size].append(np.mean(sample))

plt.figure(figsize=(12, 8))

for size in sample_sizes:
    sns.histplot(sample_means[size], kde=True, label=f'Sample Size = {size}', stat='density')

plt.title('Distribution of Sample Means for Different Sample Sizes')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.legend()

plt.show()
```
